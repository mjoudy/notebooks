### Corelation

Correlation vs. Covariance

```python
x = np.array([1, 2.2, 2.8, 4.5, 5.1, 6.9, 7, 8.5, 9.8, 10.2])
y = np.array([1, 1.5, 3, 4.1, 3.8, 5, 7.1, 6.7, 3, 5])

x1 = x - np.mean(x)
y1 = y - np.mean(y)

x2 = (x - np.mean(x))/np.var(x)
y2 = (y - np.mean(y))/np.var(y)

xy = x*y
xy11 = x1*y1
xy22 = x2*y2
cov = np.mean(xy11)
cov

covv = np.cov(x, y)
covv

cor = np.mean(xy22)
corr = np.corrcoef(x2, y2)
```

```python
plt.figure()
plt.plot(x, y, '*-')
plt.plot(x1, y1, '.-')
plt.plot(x2, y2, '.')
```
